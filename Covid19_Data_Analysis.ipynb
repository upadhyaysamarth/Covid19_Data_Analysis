```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots look better
plt.style.use('ggplot')
```

```python
# Load dataset directly from URL
url = "https://catalog.ourworldindata.org/garden/covid/latest/compact/compact.csv"
df = pd.read_csv(url)

# Show first 5 rows
df.head()
```

```python
# Check dataset info
df.info()

# Check missing values
df.isnull().sum()

# Basic statistics
df.describe()
```

```python
df = df.rename(columns={
    "Entity": "country",
    "Date": "date",
    "Total_cases": "total_cases",
    "New_cases": "new_cases",
    "Total_deaths": "total_deaths",
    "New_deaths": "new_deaths",
    # add more if needed
})

df.head()
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.isnull().sum()
```

```python
df = df.fillna(0)  # fill missing numbers with 0
```

```python
df = df.dropna()  # drop rows with blanks
```

```python
india = df[df['country'] == "India"].copy()
india.tail()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(india['date'], india['total_cases'], color="blue")
plt.title("Total COVID-19 Cases (India)")
plt.xlabel("Date")
plt.ylabel("Total Cases")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.bar(india['date'], india['new_cases'], color="orange")
plt.title("Daily New Cases (India)")
plt.xlabel("Date")
plt.ylabel("New Cases")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(india['date'], india['total_deaths'], color="red")
plt.title("Total COVID-19 Deaths (India)")
plt.xlabel("Date")
plt.ylabel("Total Deaths")
plt.xticks(rotation=45)
plt.show()
```

```python
usa = df[df['country'] == "United States"].copy()

plt.figure(figsize=(12,6))
plt.plot(india['date'], india['total_cases'], label="India")
plt.plot(usa['date'], usa['total_cases'], label="USA")
plt.legend()
plt.title("Total Cases: India vs USA")
plt.show()
```

```python
import seaborn as sns

corr = india[['total_cases', 'total_deaths', 'new_cases']].corr()

plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation (India)")
plt.show()
```
